# Reporte General de Encuesta de Ingeniería

Este notebook consolida los cinco análisis realizados sobre el archivo `encuesta_ingenieria_10000_respuestas.csv`. Cada sección lee la columna correspondiente y produce su conteo.

```python
import csv

CSV = "../encuesta_ingenieria_10000_respuestas.csv"

def leer_csv(ruta):
    """Devuelve todas las filas del CSV sin encabezado."""
    with open(ruta, "r", encoding="utf-8") as f:
        lector = csv.reader(f)
        next(lector)
        return list(lector)
```

---
## Reporte 1 — Total de encuestados
**Columna:** índice 0 (ID del encuestado).  
Se cuenta cada fila que tenga un ID no vacío.

```python
filas = leer_csv(CSV)

total_encuestados = sum(1 for fila in filas if fila[0])

print(f"Total de encuestados: {total_encuestados}")
```

---
## Reporte 2 — Estudiantes por carrera
**Columna:** índice 1 (carrera).  
Se agrupa por las tres ingenierías disponibles.

```python
carreras = ["Ingeniería Eléctrica", "Ingeniería de Sistemas", "Ingeniería Electrónica"]
conteo_carreras = {c: 0 for c in carreras}

for fila in filas:
    carrera = fila[1].strip()
    if carrera in conteo_carreras:
        conteo_carreras[carrera] += 1

for carrera, total in conteo_carreras.items():
    print(f"{carrera}: {total}")
```

---
## Reporte 3 — Estudiantes por semestre
**Columna:** índice 2 (semestre).  
Se convierte a entero y se filtra para los semestres 5, 6 y 7.

```python
semestres = {5: 0, 6: 0, 7: 0}

for fila in filas:
    sem = int(fila[2].strip())
    if sem in semestres:
        semestres[sem] += 1

for sem, total in semestres.items():
    print(f"Semestre {sem}: {total}")
```

---
## Reporte 4 — Estudiantes por jornada
**Columna:** índice 3 (jornada).  
Se clasifica en Matutina, Vespertina y Nocturna.

```python
jornadas = ["Matutina", "Vespertina", "Nocturna"]
conteo_jornadas = {j: 0 for j in jornadas}

for fila in filas:
    jornada = fila[3].strip()
    if jornada in conteo_jornadas:
        conteo_jornadas[jornada] += 1

for jornada, total in conteo_jornadas.items():
    print(f"{jornada}: {total}")
```

---
## Reporte 5 — Estudiantes que trabajan
**Columna:** índice 4 (trabaja).  
Se separa entre quienes respondieron Sí y quienes respondieron No, calculando el porcentaje de cada grupo.

```python
trabajan     = sum(1 for fila in filas if fila[4].strip() == "Sí")
no_trabajan  = sum(1 for fila in filas if fila[4].strip() == "No")
total        = trabajan + no_trabajan

print(f"Trabajan    : {trabajan}  ({trabajan / total * 100:.2f}%)")
print(f"No trabajan : {no_trabajan}  ({no_trabajan / total * 100:.2f}%)")
```
